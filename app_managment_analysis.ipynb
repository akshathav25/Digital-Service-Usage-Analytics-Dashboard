```python
import pandas as pd

df = pd.read_csv("/content/archive (3).zip")
df.info()
```

```python
df['published_at'] = pd.to_datetime(df['published_at'])
df.info()
```

```python
top_views = df.sort_values(by='view_count', ascending=False).head(10)

top_views[['Title','view_count']]
```

```python
top_engagement = df.sort_values(by='engagement_rate', ascending=False).head(10)

top_engagement[['Title','engagement_rate']]
```

```python
import matplotlib.pyplot as plt

top_views = df.sort_values(by='view_count', ascending=False).head(10)

plt.figure(figsize=(10,5))
plt.barh(top_views['Title'], top_views['view_count'])

plt.title("Top 10 Most Viewed Videos")
plt.xlabel("Views")

plt.gca().invert_yaxis()
plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(df['view_count'], df['engagement_rate'])

plt.xlabel("Views")
plt.ylabel("Engagement Rate")
plt.title("Views vs Engagement")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(df['duration_seconds'], df['engagement_rate'])

plt.xlabel("Video Duration (seconds)")
plt.ylabel("Engagement Rate")

plt.title("Video Length vs Engagement")

plt.show()
```
